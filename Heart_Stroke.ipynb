```python
class Encoding():
    def __init__(self, data):
        self.data = data
    def one_hot(self):
        from sklearn.preprocessing import OneHotEncoder
        for i in self.data.select_dtypes(include = ["object"]):
            ohe = OneHotEncoder()
            self.data[i] = ohe.fit_transform(self.data[i])
        return self.data
    def label_encoding(self):
        from sklearn.preprocessing import LabelEncoder
        for i in self.data.select_dtypes(include = ["object"]):
            enco = LabelEncoder()
            self.data[i] = enco.fit_transform(self.data[i])
    
```

```python
## features & target
x = df.drop("stroke", axis = 1)
y = df["stroke"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
lsr = LogisticRegression()
lsr.fit(x_train, y_train)
```

```python
y_pred_test = lsr.predict(x_test)
```

```python
## metrics
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
print("accuracy Score: ", accuracy_score(y_test, y_pred_test))
print("Confusion Matrix: ")
print(confusion_matrix(y_test, y_pred_test))
```

```python
## oversampling
from imblearn.over_sampling import RandomOverSampler
ovr = RandomOverSampler()
x_resampled, y_resampled = ovr.fit_resample(x,y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_resampled,y_resampled, test_size = 0.2, random_state = 42)
lsr2 = LogisticRegression()
lsr2.fit(x_train, y_train)
```

```python
y_pred_test2 = lsr2.predict(x_test)
```

```python
print("accuracy Score: ", accuracy_score(y_test, y_pred_test2))
print("Confusion Matrix: ")
print(confusion_matrix(y_test, y_pred_test2))
```
